```python
movie = pd.read_csv("dataset.csv")
```

```python
movies=movie[['id', 'title', 'overview', 'genre']]
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genre']
```

```python
movies['tags'].isnull().sum()
```

```python
new_data  = movies.drop(columns=['overview', 'genre'])
```

```python
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vector = CountVectorizer(max_features=10000 , stop_words="english")
```

```python
vector
```

```python
cv=vector.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(cv)
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda cv:cv[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
